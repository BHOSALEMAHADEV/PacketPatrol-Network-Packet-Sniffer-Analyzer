```python
!pip install scapy
from scapy.all import sniff, IP, TCP, UDP
import pandas as pd
from datetime import datetime

# List to store packet logs
packet_log = []

# Packet handler
def process_packet(packet):
    log = {}
    log['Timestamp'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    if IP in packet:
        ip_layer = packet[IP]
        log['Source IP'] = ip_layer.src
        log['Destination IP'] = ip_layer.dst
        log['Protocol'] = ip_layer.proto

        if TCP in packet:
            tcp_layer = packet[TCP]
            log['Source Port'] = tcp_layer.sport
            log['Destination Port'] = tcp_layer.dport
            log['Flags'] = tcp_layer.flags
        elif UDP in packet:
            udp_layer = packet[UDP]
            log['Source Port'] = udp_layer.sport
            log['Destination Port'] = udp_layer.dport
        else:
            log['Source Port'] = 'N/A'
            log['Destination Port'] = 'N/A'
            log['Flags'] = 'N/A'
    else:
        log['Source IP'] = 'N/A'
        log['Destination IP'] = 'N/A'
        log['Protocol'] = 'Unknown'

    packet_log.append(log)
    print(log)

# Start sniffing
def start_sniffing(interface="eth0", packet_count=10):
    print(f"[*] Starting Packet Sniffing on {interface}...")
    sniff(iface=interface, prn=process_packet, count=packet_count, store=0)
    save_logs()

# Save captured data to CSV
def save_logs():
    df = pd.DataFrame(packet_log)
    df.to_csv("packet_logs.csv", index=False)
    print("\n[+] Logs saved to packet_logs.csv")

if __name__ == "__main__":
    iface = input("Enter your network interface (e.g., eth0, wlan0): ")
    pkt_count = int(input("Enter number of packets to capture: "))
    start_sniffing(interface=iface, packet_count=pkt_count)
```
